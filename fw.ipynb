__5.С использованием t-test статистики проверить взаимосвязи между несколькими переменными.___

Так как, мы нашли несколько зависимостей на корр матрице, и они плохо видны на пэйрплоте, попробуем проверить их двусторонним t-testом. А так же проверим односторонним t-testом гипотезу, что ушла половина сотрудников.

Ho(ушла половина сотрудников) M(совокупности) = 0.5

```python
emp_left = df.left
stats.ttest_1samp(emp_left, .5)
```

Думаю, мы можем смело отклонять данную гипотезу. Ушло гораздо меньше сотрудников

Ho(удовлетворенность тех кто получал повышение = удовлетворенности тех, кто не получал) M(удвл с повышением)=M(удвл без повышения)

```python
emp_happy_1 = df[df['promotion_last_5years'] == 1]['satisfaction_level']
emp_happy_2 = df[df['promotion_last_5years'] == 0]['satisfaction_level']
stats.ttest_ind(emp_happy_1, emp_happy_2)
```

Очевидно, что гипотезу нет смысла крутить дальше! Те, что с повышением посчастливее будут!

Ho(инцидент на работе не зависит от количества времени на работе) M(рабочего времени с инцидентом)=M(рабочего времени без инцидента)

```python
with_inc = df[df['Work_accident'] == 1]['average_montly_hours']
without_inc = df[df['Work_accident'] == 0]['average_montly_hours']
stats.ttest_ind(with_inc, without_inc)
```

Ха! Принимаем гипотезу! Инциденты от кривых рук, а не от количество рабочих часов!

__10.Проверить несколько гипотез для данных,чтобы понять почему сотрудники покидают компанию.__

Проверм пару гипотез с помощью т тестов, на матрице корреляции четко видно что те кто ушел, меньше удовлетворены! А так же проверим, зависит ли повышение и количество проектов на уход сотрудников!

```python
df_predateli = df[df['left'] == 1]
df_vernie = df[df['left'] == 0]
```

Гипотеза 1. Мат ожидание радости тех кто ушел = Мат ожидание радости тех кто остался

```python
radost_pred = df_predateli.satisfaction_level
radost_vern = df_vernie.satisfaction_level
stats.ttest_ind(radost_pred, radost_vern)
```

ЧТД. Радость у них разная.

Гипотеза 2. Мат ожидание повышения тех кто ушел = Мат ожидание повышения тех кто остался

```python
vverh_pred = df_predateli.promotion_last_5years
vverh_vern = df_vernie.promotion_last_5years
stats.ttest_ind(vverh_pred, vverh_vern)
```

С таким значением p-value можно смело утверждать, что повышение сыграло свою роль!

Гипотеза 3. Мат ожидание количества проектов тех кто ушел = Мат ожидание количества проектов тех кто остался

```python
proj_pred = df_predateli.number_project
proj_vern = df_vernie.number_project
stats.ttest_ind(proj_pred, proj_vern)
```

Значение p-value на грани. Но я сомневаюсь, что количество проектов сильно повлияло на увольнение. Не буду отвергать гипотезу!

```python
df_predateli.number_project.describe()
```

```python
df_vernie.number_project.describe()
```

Да! определенно гипотеза имеет место быть!
