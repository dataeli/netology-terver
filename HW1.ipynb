Вопрос 2. Графики!

(a) Распределение Пуассона

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from scipy.stats import poisson
%matplotlib inline
```

```python
pois_rv = sts.poisson(5)
x = np.linspace(0,10,100)
cdf = pois_rv.cdf(x)
plt.step(x, cdf)

plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.title('Распределение Пуассона')
```

```python
x = np.linspace(0,50,51)
for N in [5, 10, 13, 20]:
    
    rv = poisson(N)
    pmf = rv.pmf(x)
    plt.plot(x, pmf, 'o', label="$N=%s" % (N))
plt.legend()
plt.title("PMF (Poisson)")

plt.ylabel('$F(X)$')
plt.xlabel('$x$')
```

(b) Экспоненциальное распределение

```python
exp_rv = sts.expon(5)
x = np.linspace(5,10,100)
cdf = exp_rv.cdf(x)
plt.plot(x, cdf)

plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.title('Экспоненциальное распределение')
```

```python
x = np.linspace(0,25,26)
for N in [2, 9, 15, 20]:
    
    rv = sts.expon(N)
    pdf = rv.pdf(x)
    plt.plot(x, pdf, label="$N=%s" % (N))
plt.legend()
plt.title("PDF (Expon)")

plt.ylabel('$F(X)$')
plt.xlabel('$x$')
```

(с) Распределение Коши

```python
cau_rv = sts.cauchy(5)
x = np.linspace(0,10,100)
cdf = cau_rv.cdf(x)
plt.plot(x, cdf)

plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.title('CAUCHY')
```

```python
x = np.linspace(0,25,26)
for N in [2, 9, 15, 20]:
    
    rv = sts.cauchy(N)
    pdf = rv.pdf(x)
    plt.plot(x, pdf, label="$N=%s" % (N))
plt.legend()
plt.title("PDF (CAUCHY)")

plt.ylabel('$F(X)$')
plt.xlabel('$x$')
```
